# Extension and validation of coordinates was implemented in `simex` and next cells will call cli `extend_metadata_of_files`

Get dirs that hold WAV, JPG and AVI files

```python
%%bash
dir_to_be_processed="/LUSTRE/sacmod/SIPECAM/data/"
if [ -f ~/list_dirs.txt ]; then
     rm ~/list_dirs.txt
fi
if [ -f ~/list_dirs_auxiliary.txt ]; then
     rm ~/list_dirs_auxiliary.txt
fi
SAVEIFS=$IFS;IFS=$(echo -en "\n\b")
for f in $(find "$dir_to_be_processed" -name "*.WAV" -o -name "*.JPG" -o -name "*.AVI")
  do
     direc=$(dirname $f)
     echo "$direc" >> ~/list_dirs_auxiliary.txt
done
direcs=$(cat ~/list_dirs_auxiliary.txt|uniq)
echo "$direcs" > ~/list_dirs.txt
IFS=$SAVEIFS
```

```python
%%bash
SAVEIFS=$IFS;IFS=$(echo -en "\n\b")
for d in $(cat ~/list_dirs.txt)
do
  json=$(find "$d" -name "*.json")
  echo "$json"
  echo "------------"
  echo "------------"
done
IFS=$SAVEIFS
```
